```python
class Person:
  def __init__(self,name,age,gender,salary,department,remote):
    self.name = name
    self.age = age
    self.gender = gender
    self.salary = salary
    self.department = department
    self.remote = remote
  def inperson(self):
    print(self.name)
    print(self.age)
    print(self.gender)
    print(self.salary)
    print(self.department)
    print(self.remote)





```

```python
o = Person("Sneha",22,"Female",100000,"Electronics and Telecommunication",True)
```

```python
o.inperson()
```
